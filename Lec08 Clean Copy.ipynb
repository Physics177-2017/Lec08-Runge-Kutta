```python
from math import sin
import numpy as np
import matplotlib.pyplot as plt # different from textbook

def f(x,y):
    return -x**3 + sin(t)

a = 0.0
b = 10.0
N = 10
dt = (b-a)/N

tpoints = np.arange(a,b,dt)
xpoints = []
x0 = 0.0

#########################

x=x0
for t in tpoints:
    xpoints.append(x)
    k1 = dt*f(x,t)
    k2 = dt*f(x+ 0.5*k1, t + 0.5*dt)
    x += k2
    

#########################

xx = np.array(xpoints)
tt = np.array(tpoints)

print(xx)
print(tt)

plt.plot(tt,xx)
plt.show()
```

## Fourth Order

```python
from math import sin
import numpy as np
import matplotlib.pyplot as plt # different from textbook

def f(x,y):
    return -x**3 + sin(t)

a = 0.0
b = 10.0
N = 10
dt = (b-a)/N

tpoints = np.arange(a,b,dt)
xpoints = []
x0 = 0.0

#########################

x=x0
for t in tpoints:
    xpoints.append(x)
    k1 = dt*f(x,t)
    k2 = dt*f(x+ 0.5*k1, t + 0.5*dt)
    k3 = dt*f(x+ 0.5*k2, t + 0.5*dt) ## NEW
    k4 = dt*f(x+k3, t+dt) ## NEW
    x += (k1 + 2*k2 + 2*k3 + k4)/6.
    

#########################

xx4 = np.array(xpoints)
tt4 = np.array(tpoints)


plt.plot(tt,xx, label='RK2')
plt.plot(tte,xxe, label='Eul')
plt.plot(tt4,xx4, label='RK4')
plt.plot(tt2,xx2, label='RK2_100')
plt.legend()
plt.show()
```
